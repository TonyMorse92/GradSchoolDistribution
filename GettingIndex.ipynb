```python
import requests
from bs4 import BeautifulSoup

ssl_verify_path = 'ch.pem'

url = f'https://mathgenealogy.org/index.php'
page = requests.get(url,verify=ssl_verify_path)

soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
name = soup.p
num = name.find_next('p')
print(num.prettify())
print(num.text.strip())
```

```python
name = soup.p
num = name.find_next('p')
print(num.prettify())
print(num.text.strip())
print(num.text.strip().rsplit()[0])
```

```python
print("\n\n", "Number records: ", num.text.strip().rsplit()[0])
```
